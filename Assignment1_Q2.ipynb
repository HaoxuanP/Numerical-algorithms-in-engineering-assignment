```python
from collections import deque

# Divide 1 priority queue into 3 queues with different levels
L1 = deque()
L2 = deque()
L3 = deque()

# Jobs to be printed
output = []

# Number of jobs submitted
submit_count = 0

# Open input file (Change file name here for different test cases)
with open('q2_in_8.txt', 'r') as q2_in:

    N = int(q2_in.readline()) # total no. ops
    p = int(q2_in.readline()) # no. submit
    q = int(q2_in.readline()) # no. print

    for line in range(N):
          line = q2_in.readline().strip().split()

          # if submit job (1)
          if line[0] == '1':
              level = line[1]
              value = int(line[2])
              submit_count += 1

              if level == 'L1':
                  L1.append(value)
              elif level == 'L2':
                  L2.append(value)
              elif level == 'L3':
                  L3.append(value)

          # if print job (2)
          elif line[0] == '2':
              if L1:
                  output.append(L1.popleft())
              elif L2:
                  output.append(L2.popleft())
              elif L3:
                  output.append(L3.popleft())

          # if empty
          else:
              raise Exception('The queue is empty!')

# Check if the number of print and submit are consistent with input file
if submit_count != p:
    raise Exception('Incorrect number of jobs submited')


if len(output) != q:
    raise Exception('Incorrect number of jobs for print!')




# Write answers as output file
with open('q2_out.txt', 'w') as q2_out:
    for element in output:
        q2_out.write(f"{element}\n")

```
