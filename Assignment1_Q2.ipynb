```python
from collections import deque

class PrintQueues:
    def __init__(self):
        self.queues = {"L1": deque(),
                       "L2": deque(),
                       "L3": deque()}

    def submit(self, level, value):
        self.queues[level].append(value)

    def print_one_job(self):
        for level in ["L1", "L2", "L3"]:
            if self.queues[level]:
                return self.queues[level].popleft()



printer = PrintQueues()
output = []

# Open input file (Change file name here for different test cases)
with open('q2_in_10.txt', 'r') as q2_in:

    N = int(q2_in.readline()) # total no. ops
    p = int(q2_in.readline()) # no. submit
    q = int(q2_in.readline()) # no. print

    for line in range(N):
          job = q2_in.readline().strip().split()

          # if submit job (1)
          # append jobs to corresponding level
          if job[0] == '1':
              level = job[1]
              value = job[2]
              printer.submit(level,value)

          # if print job (2)
          # print top priority jobs
          elif job[0] == '2':
              out = printer.print_one_job()
              output.append(out)

# Write answers as output file
with open('q2_out.txt', 'w') as q2_out:
    for element in output:
        q2_out.write(f"{element}\n")
```
